```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv(r'D:/Malathi/SEM_6/IVP/class.csv')
data
```

```python
data.info()
```

```python
plt.hist(data['Height'], bins=5, edgecolor='black')
plt.xlabel('Height (mts)')
plt.ylabel('Number of Students')
plt.title('Height Distribution of Students')
plt.show()
```

```python
plt.hist(data['Weight'], bins=5, edgecolor='black')
plt.xlabel('Weight (kgs)')
plt.ylabel('Number of Students')
plt.title('Weight Distribution of Students')
plt.show()
```

```python
sns.scatterplot(x='Height', y='Weight', data=data, color='blue', alpha=0.7)
plt.xlabel('Height (ft)')
plt.ylabel('Weight (lbs)')
plt.title('Joint Distribution of Height and Weight')
plt.show()
```

```python
heatmap_data = data[["Height", "Weight"]]
sns.jointplot(heatmap_data, x='Height', y='Weight')
plt.show()
```

```python
mean_height = np.mean(data['Height'])
mean_weight = np.mean(data['Weight'])
```

```python
median_height = np.median(data['Height'])
median_weight = np.median(data['Weight'])
```

```python
mode_height = stats.mode(data['Height'])
mode_weight = stats.mode(data['Weight'])
```

```python
std_height = np.std(data['Height'])
std_weight = np.std(data['Weight'])
```

```python
var_height = np.var(data['Height'])
var_weight = np.var(data['Weight'])
```

```python
skew_height = stats.skew(data['Height'])
skew_weight = stats.skew(data['Weight'])
```

```python
kurtosis_height = stats.kurtosis(data['Height'])
kurtosis_weight = stats.kurtosis(data['Weight'])
```

```python
print("Height:")
print("Mean:", mean_height)
print("Median:", median_height)
print("Mode:", mode_height)
print("Standard Deviation:", std_height)
print("Variance:", var_height)
print("Skewness:", skew_height)
print("Kurtosis:", kurtosis_height)

print("\nWeight:")
print("Mean:", mean_weight)
print("Median:", median_weight)
print("Mode:", mode_weight)
print("Standard Deviation:", std_weight)
print("Variance:", var_weight)
print("Skewness:", skew_weight)
print("Kurtosis:", kurtosis_weight)
```
