```python
import numpy as np
import matplotlib.pyplot as plt

def plot_uniform_distribution(lower_bound, upper_bound, num_points=1000):
    # Generate x values
    x = np.linspace(lower_bound, upper_bound, num_points)
    # Compute the probability density function (PDF)
    pdf = np.ones_like(x) / (upper_bound - lower_bound)


    return x, pdf

# Parameters
lower_bound = 0  # Lower bound of the range
upper_bound = 10  # Upper bound of the range

# Plot the uniform distribution PDF
x, pdf = plot_uniform_distribution(lower_bound, upper_bound)
# Plot the PDF
plt.plot(x, pdf, label=f'Uniform({lower_bound}, {upper_bound})')
plt.title('Uniform Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)    
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_uniform_cdf(lower_bound, upper_bound, num_points=1000):
    # Generate x values
    x = np.linspace(lower_bound, upper_bound, num_points)

    # Compute the cumulative distribution function (CDF)
    cdf = (x - lower_bound) / (upper_bound - lower_bound)
    cdf[x < lower_bound] = 0
    cdf[x > upper_bound] = 1

    return x, cdf


# Parameters
lower_bound = 0  # Lower bound of the range
upper_bound = 10  # Upper bound of the range

# Plot the uniform distribution CDF
x, cdf = plot_uniform_cdf(lower_bound, upper_bound)
plt.plot(x, cdf, label=f'Uniform({lower_bound}, {upper_bound})')
plt.title('Uniform Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)
plt.show()
```
